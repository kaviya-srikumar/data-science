```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("Amazon Sale Report.csv")
df.head()
```

```python
print("Shape of Dataset:", df.shape)
print("\nColumn Names:\n", df.columns)
print("\nDataset Info:\n")
df.info()

print("\nMissing Values:\n")
print(df.isna().sum())

```

```python
df.describe(include="all")
```

```python
print("Total duplicate rows:", df.duplicated().sum())
df = df.drop_duplicates()
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    plt.figure(figsize=(6,4))
    plt.hist(df[col].dropna(), bins=30,edgecolor="black",color="Skyblue")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
sns.pairplot(df.select_dtypes(include=['int64','float64']))
plt.show()

```

```python
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(df[col])
    plt.title(f"Outliers in {col}")
    plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    if df[col].nunique() < 15:   # prevents huge plots
        plt.figure(figsize=(8,4))
        sns.countplot(y=df[col], order=df[col].value_counts().index)
        plt.title(f"Count Plot of {col}")
        plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='ignore')

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()

df.groupby('Month').size().plot(kind='bar', figsize=(10,4))
plt.title("Orders per Month")
plt.show()
```

```python
if 'Category' in df.columns and 'Amount' in df.columns:
    df.groupby('Category')['Amount'].sum().sort_values().plot(kind='bar')
    plt.title("Sales by Category")
    plt.ylabel("Total Sales")
    plt.show()
```
