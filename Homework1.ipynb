```python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(0, 11, 1)


y = x**2

plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o', linestyle='--',color="red")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("X Vs Y plot")
plt.grid(True)
plt.show()

```

```python


x = ['A', 'B', 'C', 'D']
y = [40, 70, 30, 90]


plt.figure(figsize=(6, 4))
plt.bar(x, y, color='Pink')
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Product-sales")
plt.grid(True)
plt.show()
```

```python
x = np.arange(1, 11)
y = x ** 3

plt.figure(figsize=(5, 7))

plt.subplot(2, 2, 1)
plt.plot(x, y, marker='o',color="black")
plt.title("Line Plot")
plt.grid(True)

plt.subplot(2, 2, 3)
plt.scatter(x, y,color="black")
plt.title("Scatter Plot")
plt.grid(True)

plt.subplot(2, 2, 2)
plt.bar(['A','B','C','D'], [40, 70, 30, 90],color="black")
plt.title("Bar Plot")
plt.grid(True)

plt.subplot(2, 2, 4)
plt.hist(np.random.randn(100),color="black")
plt.title("Histogram")
plt.grid(True)


plt.show()

```

```python

df = np.random.randn(200)

plt.figure(figsize=(6, 4))
plt.hist(df, bins=20, edgecolor='black')
plt.title("Histogram of data")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.grid(True)
plt.show()
```
