```python
s=int(input())
f=[]
for j in range(s):
  num=int(input())
  f.append(num)
t=list(map(int,input().split()))
print(f)
print(t)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(2,2))
plt.plot(f,t)

plt.xlabel('days')
plt.ylabel('sales')
plt.title('sales-plot')
plt.show()
```

```python
import numpy as np
import pandas as pd

np.random.seed(40)

data={
    "x" : np.arange(1,10),
    "y" : np.random.randint(10,100,size=9)
}
df=pd.DataFrame(data)
print(df)

plt.figure(figsize=(2,2))
plt.plot(df)
plt.xlabel("price")
plt.ylabel("quantity")
plt.title("sales-plot")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.plot(df["x"],df["y"],marker="o")
plt.xlabel("price")
plt.ylabel("quantity")
plt.title("sales-plot")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.plot(df["x"],df["y"],marker="o",color="pink")
plt.xlabel("price")
plt.ylabel("quantity")
plt.title("sales-plot")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd

np.random.seed(60)

data={
    "x" : np.arange(1,21),
    "y" : np.random.randint(100,1000,size=20)
}
df=pd.DataFrame(data)
print(df)
plt.figure(figsize=(4,4))
plt.plot(df["x"],df["y"],marker="o",color="grey")
plt.xlabel("price")
plt.ylabel("quantity")
plt.title("sales-plot")
plt.grid(True)
plt.show()
```

```python
import numpy as np
x=np.arange(1,11)
y=np.random.randint(10,100,size=10)
data=np.random.randn(100)

print(x)
print(y)
```

```python
plt.figure(figsize=(4,4))
plt.plot(x,y,marker='x',linestyle='--',color="grey")
plt.xlabel("price")
plt.ylabel("quantity")
plt.title("sales-plot")
plt.grid(True)
plt.show()
```
